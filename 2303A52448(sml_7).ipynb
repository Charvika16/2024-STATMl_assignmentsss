Question 1:

1.Read the data with pandas and find features and target variables

2.Plot a graph between features and target

3.Find Best fit line using linear regression.

4.FInd MSE, MAE, for test size (20,25)

```python
import pandas as pd
```

```python
A=pd.read_csv('/content/Salary_Data (1).csv')
print(A)
```

```python
y=A['Salary']
print(y)
```

```python
x=A.drop('Salary',axis=1)
print(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.scatter(A['YearsExperience'],A['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('YearsExperience vs Salary')
plt.show()
```

```python
x_mean = np.mean(A['YearsExperience'])
y_mean = np.mean(A['Salary'])

numerator = np.sum((A['YearsExperience'] - x_mean) * (A['Salary'] - y_mean))
denominator = np.sum((A['YearsExperience'] - x_mean) ** 2)

slope = numerator / denominator

intercept = y_mean - (slope * x_mean)

print("Slope:", slope)
print("Intercept:", intercept)

plt.scatter(A['YearsExperience'], A['Salary'])
plt.plot(A['YearsExperience'], (slope * A['YearsExperience'] + intercept), color='green')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('YearsExperience vs Salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=34)

y_pred = (slope * X_test['YearsExperience'] + intercept)


mse = np.mean((y_test - y_pred) ** 2)
print("Mean Squared Error (MSE):", mse)


mae = np.mean(np.abs(y_test - y_pred))
print("Mean Absolute Error (MAE):", mae)


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=34)


y_pred = (slope * X_test['YearsExperience'] + intercept)


mse = np.mean((y_test - y_pred) ** 2)
print("Mean Squared Error (MSE) for test size 0.25:", mse)


mae = np.mean(np.abs(y_test - y_pred))
print("Mean Absolute Error (MAE) for test size 0.25:", mae)
```
