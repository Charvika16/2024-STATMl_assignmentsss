```python
import pandas as pd
```

```python
A=pd.read_csv('/content/diabetes_data_upload.csv')
print(A)
```

```python
y=A['class']
print(y)
```

```python
x=A.drop('class',axis=1)
print(x)
x=x.fillna(0)
```

```python
mapping={'Positive':1,'Negative':0}
y=y.map(mapping)
print(y)
```

```python
mapping={'Male':1,'Female':0}
x['Gender']=x['Gender'].map(mapping)
print(x)
```

```python
mapping={'Yes':1,'No':0}
x=x.replace('Yes',1)
x=x.replace('No',0)
print(x)
```

```python
import matplotlib.pyplot as plt
```

```python
for column in x.columns:
  plt.plot(x[column],y)
  plt.xlabel(column)
  plt.ylabel('class')
  plt.title(f'Scatter plot of {column} vs class')
  plt.show()
```
