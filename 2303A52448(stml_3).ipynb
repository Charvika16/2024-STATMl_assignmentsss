FIND OUT THE MAXIMUM LIKELIHOOD VALUES FOR CALIFORNIA_HOUSING['LONGITUDE','LATITUDE',HOUSING_MEDIAN_AGE',TOTAL_BEDROOMS']DATA



```python
import pandas as pd
import numpy as np
A=pd.read_csv('/content/sample_data/california_housing_test.csv')
print(A)
```

```python
Column=['latitude','total_rooms','housing_median_age','total_bedrooms']
M=A['latitude'].mean()
print("Mean of the latitude:",M)
print("Mean of the total_rooms:",A['total_rooms'].mean())
print("mean of the housing_median_age:",A['housing_median_age'].mean())
print("mean of the total_bedrooms:",A['total_bedrooms'].mean())
```

```python
print("standard deviation of each column:",A.std())
```

```python
V=A['latitude'].var()
print("standard deviation of the latitude:",V)
print("standard deviation of total_rooms:",A['total_rooms'].std())
print("standard deviation of housing_median_age:",A['housing_median_age'].std())
print("standard deviation of total_bedrooms:",A['total_bedrooms'].std())
```

**MAXIMUM LIKELIHOOD ESTIMATION**

LATITUDE


```python
a= np.mean(A['latitude'])
print("Maximum Likelihood of Latitude:")
print("")
A['result']=(1/np.sqrt(2*np.pi*np.mean(A['latitude']**2))*np.log(((a-np.std(A['latitude']))**2)/(2*np.mean(A['latitude']**2))))
print(A[['latitude','result']])
```

TOTAL ROOMS

```python
print("Maximum likelihood of total_rooms:")
A['result']=(1/np.sqrt(2*np.pi*np.mean(A['total_rooms']**2))*np.log(((a-np.std(A['total_rooms']))**2)/(2*np.mean(A['total_rooms']**2))))
print(A[['total_rooms','result']])
```

HOUSING MEDIAN AGE

```python
print("Maximum likelihood of housing_median_age:")
A['result']=(1/np.sqrt(2*np.pi*np.mean(A['housing_median_age']**2))*np.log(((a-np.std(A['housing_median_age']))**2)/(2*np.mean(A['housing_median_age']**2))))
print(A[['housing_median_age','result']])
```

Question 2:

### **Find log-likelihood values for 50, 75, and 80. For California_housing [‘latitude’] and find the Max likelihood values.**

```python
ml=np.mean(A['latitude'])
s =np.var(A['latitude'],ddof=1)
```

```python
def Log_Likelihood(a,ml,si):
  return -0.5*(np.log(2*np.pi*s**2)+((a-ml)**2/s**2))
```

```python
values=[50,75,80]
Log_Likelihoods=[Log_Likelihood(v, ml,s)for v in values]
```

```python
print("Mean :",ml)
print("Varience:",s)
print("Log_Likelihood for values[50,75,80]:",Log_Likelihoods)
```

**QUESTION-3**
**Estimate the density of each column of california_housing [‘latitude’,total_rooms’,’house_median_age’,’total_bedrooms’] data, For the bins [5, 10, 15, 20]. And write the observations**.

```python
import matplotlib.pyplot as plt
columns = ['latitude','total_rooms','housing_median_age','total_bedrooms']
bins=5
```

```python
for column in columns:
  plt.figure(figsize=(9,7))
  plt.hist(A[column],bins=bins,density=True,alpha=0.5,color='green')
  plt.title(f'Density plot for the {column}')
  plt.xlabel(column)
  plt.ylabel('Density')
  plt.grid(True)
  plt.show()
```
